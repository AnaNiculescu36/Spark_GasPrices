```python
import pyspark as pyspark
import os
from pyspark import *
os.environ['SPARK_HOME'] = '/home/ana/spark/spark-2.3.2-bin-hadoop2.7'
#sc = pyspark.SparkContext.getOrCreate()
conf = SparkConf().setAppName("App")
conf = (conf.setMaster('local[*]')
        .set('spark.driver.memory', '4G'))
sc = SparkContext(conf=conf)
sc.setLogLevel("OFF")
sql_sc = pyspark.SQLContext(sc)


def get_data_from_file(file_path, sql_sc):
    data_df = (sql_sc
             .read
             .option('header', 'false')
             .option('sep', ';')
             .option('inferSchema', 'true')
             .csv(file_path))
    return data_df
```

```python
prix2014_df = get_data_from_file("gas_data/Prix2014.csv", sql_sc)
prix2015_df = get_data_from_file("gas_data/Prix2015.csv", sql_sc)
prix2016_df = get_data_from_file("gas_data/Prix2016.csv", sql_sc)
prix2017_df = get_data_from_file("gas_data/Prix2017.csv", sql_sc)

services_df = get_data_from_file("gas_data/Services2017.csv", sql_sc)
stations_df = get_data_from_file("gas_data/Stations2017.csv", sql_sc)
```

```python
all_prix_df = (prix2014_df
              .union(prix2015_df)
              .union(prix2016_df)
              .union(prix2017_df))
```

```python
# changing the column names
all_prix_df = (all_prix_df.selectExpr("_c0 as id", "_c1 as zip_code",
                              "_c2 as pop_type", "_c3 as lat",
                              "_c4 as lon", "_c5 as date",
                              "_c6 as gas_id", "_c7 as gas_label",
                              "_c8 as gas_price "))
```

```python
import pyspark.sql.functions as func
all_prix_df = (all_prix_df.
               withColumn("lat", func.round((func.col("lat")/100000), 3)).
               withColumn("lon", func.round((func.col("lon")/100000), 3)))
```

```python
from pyspark.sql.functions import year, month, dayofmonth, col
all_prix_df = (all_prix_df
            .withColumn(
            "year", year(col("date"))
            ).withColumn(
            "month", month(col("date"))
            ).withColumn(
            "day", dayofmonth(col("date"))
            ))
```

```python
from pyspark.sql import Window
w = Window.partitionBy('gas_label','day', 'month', 'year')
all_prix_df = (all_prix_df.select('*', func.avg('gas_price').
                            over(w).
                            alias('avg_day_price')))
all_prix_df =  all_prix_df.sort("date")
all_prix_df =  all_prix_df.where(col("date").isNotNull())
```

```python
all_prix_df = all_prix_df.withColumn("price_index", 100 * ((all_prix_df['gas_price']- all_prix_df['avg_day_price'])/all_prix_df['avg_day_price'] + 1))
```

```python
from pyspark.sql.functions import date_format, to_date
from pyspark.sql.functions import when
all_prix_df = (all_prix_df
               .withColumn("week_index",
                           date_format(
                           to_date("date", "yyyy-MM-dd HH:MM:SS"), "w")))

# each year has 52 weeks. Add multiples of 52 to
# provide week index with respect to the entire data set
# which spans over 4 years
all_prix_df = (all_prix_df
              .withColumn("week_index", when(col("year") == 2015, 52 + all_prix_df['week_index'])
                          .otherwise(col("week_index"))))
all_prix_df = (all_prix_df
              .withColumn("week_index", when(col("year") == 2016, 2*52 + all_prix_df['week_index'])
                          .otherwise(col("week_index"))))
all_prix_df = (all_prix_df
              .withColumn("week_index", when(col("year") == 2017, 3*52 + all_prix_df['week_index'])
                          .otherwise(col("week_index"))))
```

```python
# caluclate average weekly price index
w = Window.partitionBy('week_index', 'gas_label')
all_prix_df = (all_prix_df.select('*', func.avg('price_index').
                            over(w).
                            alias('avg_week_index')))
all_prix_df =  all_prix_df.sort("date")
```

```python
prix_2014 = all_prix_df.filter("year == 2014").toPandas()
```

```python
import pandas as pd
prix_2014["price_index"] = pd.to_numeric(prix_2014["price_index"])
prix_2014["avg_day_price"] = pd.to_numeric(prix_2014["avg_day_price"])
prix_2014.head()
```

```python
grouped_prix_2014 = (prix_2014
                    .groupby(['week_index', 'gas_label'])['price_index', 'avg_day_price']
                    .mean()
                    .reset_index())
```

```python
grouped_prix_2014["week_index"] = pd.to_numeric(grouped_prix_2014["week_index"])
sorted = grouped_prix_2014.sort_values(by='week_index', ascending=1)
```

```python
gas_station_df = all_prix_df.filter((all_prix_df.id == "1000008") & (all_prix_df.gas_id == "1"))
w = Window.partitionBy("day","month").orderBy(func.col("date"))

gas_station_df = gas_station_df.withColumn('lag1', func.lag(gas_station_df['gas_price']).over(w))
gas_station_df = gas_station_df.withColumn('lag2', func.lag(gas_station_df['lag1']).over(w))
gas_station_df = gas_station_df.withColumn('lag3', func.lag(gas_station_df['lag2']).over(w))
gas_station_df = gas_station_df.withColumn('lag4', func.lag(gas_station_df['lag3']).over(w))
gas_station_df = gas_station_df.withColumn('lag5', func.lag(gas_station_df['lag4']).over(w))
```

```python
from pyspark.ml import Pipeline
from pyspark.ml.regression import RandomForestRegressor
from pyspark.ml.feature import VectorIndexer, VectorAssembler
from pyspark.ml.evaluation import RegressionEvaluator

# Automatically identify categorical features, and index them.
# Set maxCategories so features with > 4 distinct values are treated as continuous.
featureIndexer =\
    (VectorAssembler(inputCols=["lag1", "lag2", "lag3", "lag4", "lag5"], outputCol="gas_price").fit(gas_station_df))

# Split the data into training and test sets (30% held out for testing)
(trainingData, testData) = gas_station_df.randomSplit([0.7, 0.3])
```

```python
# Train a RandomForest model.
rf = RandomForestRegressor(featuresCol="indexedFeatures")

# Chain indexer and forest in a Pipeline
pipeline = Pipeline(stages=[featureIndexer, rf])

# Train model.  This also runs the indexer.
model = pipeline.fit(trainingData)

# Make predictions.
predictions = model.transform(testData)

# Select example rows to display.
predictions.select("prediction", "label", "features").show(5)

# Select (prediction, true label) and compute test error
evaluator = RegressionEvaluator(
    labelCol="label", predictionCol="prediction", metricName="rmse")
rmse = evaluator.evaluate(predictions)
print("Root Mean Squared Error (RMSE) on test data = %g" % rmse)

rfModel = model.stages[1]
```
