## Forecasting gas prices in France
Build a model based on Spark ML to forecast the next day price for a
gas type in a station

```python
import pyspark as pyspark
import os
from pyspark import *
import requests
import zipfile
os.environ['SPARK_HOME'] = '/home/ana/spark/spark-2.3.2-bin-hadoop2.7'
#sc = pyspark.SparkContext.getOrCreate()
conf = SparkConf().setAppName("App")
conf = (conf.set("spark.driver.maxResultSize", "2g")
            .set("spark.driver.memory", "8g")
            .set("spark.driver.cores", "4")
            .set("spark.sql.broadcastTimeout", 3600)
            .setMaster('local[4]'))
sc = SparkContext(conf=conf)
sc.setLogLevel("OFF")
sql_sc = pyspark.SQLContext(sc)

from pyspark.sql import SparkSession
spark = SparkSession.builder.appName("Spark Dec 17").getOrCreate()
```

```python
# function that downloads directly from github the zip file with gas prices,
# saves it locally and reads it using the spark context
# the dataset_name argument is the plain text name of the file we are interested in
# eg. Prix20014, without extension or location
# the argument "spark" is the spark context use to read the data
# the output is a ddf containing the data read from the target github file
def download_and_load(dataset_name, spark):
    url = 'https://github.com/rvm-courses/GasPrices/blob/master/' + dataset_name + '.zip?raw=true'
    zip_path = "gas_data/" + dataset_name + ".zip"
    r = requests.get(url, allow_redirects=True)
    open(zip_path, 'wb').write(r.content)
    with zipfile.ZipFile(zip_path, "r") as zip_ref:
                   zip_ref.extractall("gas_data")
    file_path = "gas_data/" + dataset_name + ".csv"
    price_ddf = (spark
                .read
                .option('sep', ';')
                .option('inferSchema', 'true')
                .option('header', 'false')
                .csv(file_path)
                .selectExpr(
                '_c0 as id',
                '_c1 as zip_code',
                '_c2 as station_type',
                '_c3 as lat',
                '_c4 as lon',
                '_c5 as date',
                '_c6 as gas_id',
                '_c7 as gas_label',
                '_c8 / 1000 as gas_price')
                ).cache()
    return price_ddf
```

```python
# function that tries to create a spark ddf by reading 
# the local csv files. If the local csv file is not found
# then the "download_and_load" function is called to download
# the data directly from the github url
def get_data_from_file(dataset_name, spark):
    try:
        file_path = "gas_prices/" + dataset_name + ".csv"
        price_ddf = (spark
                .read
                .option('sep', ';')
                .option('inferSchema', 'true')
                .option('header', 'false')
                .csv(file_path)
                .selectExpr(
                '_c0 as id',
                '_c1 as zip_code',
                '_c2 as station_type',
                '_c3 as lat',
                '_c4 as lon',
                '_c5 as date',
                '_c6 as gas_id',
                '_c7 as gas_label',
                '_c8 / 1000 as gas_price')
                ).cache()
    except:
        price_ddf = download_and_load(dataset_name, spark)
    return price_ddf
```

## Data preparation
Read all files for prices

```python
prix2014_df = get_data_from_file("Prix2014", spark)
prix2015_df = get_data_from_file("Prix2015", spark)
prix2016_df = get_data_from_file("Prix2016", spark)
prix2017_df = get_data_from_file("Prix2017", spark)

#services_df = get_data_from_file("gas_data/Services2017.csv", spark)
#stations_df = get_data_from_file("gas_data/Stations2017.csv", spark)
```

Merge all files

```python
all_prix_df = (prix2014_df
              .union(prix2015_df)
              .union(prix2016_df)
              .union(prix2017_df).cache()
              )
```

Prepare latitude & longitude for mapping (divide by the right power of
10)

```python
import pyspark.sql.functions as func
all_prix_df = (all_prix_df.
               withColumn("lat", func.round((func.col("lat")/100000), 3)).
               withColumn("lon", func.round((func.col("lon")/100000), 3)))
```

Split date in year, month, week of the year

```python
from pyspark.sql.functions import year, month, dayofmonth, col
all_prix_df = (all_prix_df
            .withColumn(
            "year", year(col("date"))
            ).withColumn(
            "month", month(col("date"))
            ).withColumn(
            "day", dayofmonth(col("date"))
            ))
```

Calculate average day price in France for each gas type.

```python
from pyspark.sql import Window
w = Window.partitionBy('gas_label','day', 'month', 'year')
all_prix_df = (all_prix_df.select('*', func.avg('gas_price').
                            over(w).
                            alias('avg_day_price')))
all_prix_df =  all_prix_df.sort("date")
all_prix_df =  all_prix_df.where(col("date").isNotNull())
```

Calculate price index.

```python
all_prix_df = (all_prix_df
               .withColumn("price_index", 100 * ((all_prix_df['gas_price']- all_prix_df['avg_day_price'])/all_prix_df['avg_day_price'] + 1)))
```

Add week index

```python
from pyspark.sql.functions import date_format, to_date
from pyspark.sql.functions import when
from pyspark.sql.types import IntegerType
all_prix_df = (all_prix_df
               .withColumn("week_index",
                           date_format(
                           to_date("date", "yyyy-MM-dd HH:MM:SS"), "w")))

# each year has 52 weeks. Add multiples of 52 to
# provide week index with respect to the entire data set
# which spans over 4 years
all_prix_df = (all_prix_df
              .withColumn("week_index", when(col("year") == 2015, 52 + all_prix_df['week_index'])
                          .otherwise(col("week_index"))))
all_prix_df = (all_prix_df
              .withColumn("week_index", when(col("year") == 2016, 2*52 + all_prix_df['week_index'])
                          .otherwise(col("week_index"))))
all_prix_df = (all_prix_df
              .withColumn("week_index", when(col("year") == 2017, 3*52 + all_prix_df['week_index'])
                          .otherwise(col("week_index"))))
# cast week index to integer to be able to properly sort later
all_prix_df = (all_prix_df
               .withColumn("week_index", all_prix_df["week_index"]
                           .cast(IntegerType())))
```

```python
# caluclate average weekly price index
plotdata_df_week = (all_prix_df
    .dropna(subset=['gas_label', 'date', 'gas_price', 'week_index', 'price_index'])
    .groupBy(['week_index', 'gas_label'])
    .agg(func.mean('gas_price').alias('price'),
        func.mean('price_index').alias('price_index'))
    .orderBy(['week_index', 'gas_label'])
    .filter(func.col('gas_label').isin(['Gazole', 'SP98', 'E10', 'SP95']))
    ).toPandas()
```

```python
gas_station_df = (all_prix_df
    .dropna(subset=['gas_label', 'date', 'gas_price', 'day', 'month'])
    .orderBy(['date'])
    .filter(func.col('gas_label').isin(['Gazole']))
)
```

```python
# add lags used in random forest regressor
w = Window.partitionBy("day").orderBy(func.col("day"))

gas_station_df = gas_station_df.withColumn('lag1', func.lag(gas_station_df['gas_price']).over(w))
gas_station_df = gas_station_df.withColumn('lag2', func.lag(gas_station_df['gas_price'], 2).over(w))
gas_station_df = gas_station_df.withColumn('lag3', func.lag(gas_station_df['gas_price'], 3).over(w))
gas_station_df = gas_station_df.withColumn('lag4', func.lag(gas_station_df['gas_price'], 4).over(w))
gas_station_df = gas_station_df.withColumn('lag5', func.lag(gas_station_df['gas_price'], 5).over(w))

gas_station_df = (gas_station_df
    .dropna(subset=['lag1', 'lag2', 'lag4', 'lag5', 'lag3', 'gas_price'])
    .orderBy(['date']))
```

```python
from pyspark.ml import Pipeline
from pyspark.ml.regression import RandomForestRegressor
from pyspark.ml.feature import StringIndexer, VectorAssembler, OneHotEncoderEstimator
(trainingData, testData) = gas_station_df.randomSplit([0.7, 0.3])
```

```python
numericCols = ['lag1', 'lag2', 'lag3', 'lag4', 'lag5']
assemblerInputs =  numericCols
assembler = VectorAssembler(inputCols=assemblerInputs, outputCol="features")
```

```python
# Train a RandomForest model.
rf = RandomForestRegressor(labelCol="gas_price", featuresCol="features")
```

```python
# Chain assembler and forest in a Pipeline
pipeline = Pipeline(stages=[assembler, rf])
```

```python
model = pipeline.fit(trainingData)
```

```python
# Make predictions.
predictions = model.transform(testData)

predictions.select([c for c in predictions.columns if c in ['week_index', 'gas_label' ,'prediction']]).show(8)
```

```python
from pyspark.ml.evaluation import RegressionEvaluator
# Select (prediction, true label) and compute test error
evaluator = RegressionEvaluator(
    labelCol="gas_price", predictionCol="prediction", metricName="rmse")
rmse = evaluator.evaluate(predictions)
print("Root Mean Squared Error (RMSE) on test data = %g" % rmse)
#print("Mean Squared Error (MSE) on test data = %g" % mse)

rfModel = model.stages[1]
print(rfModel)  # summary only
```
